```python
def is_leap(year):
    leap = False                            # Initially set "leap" to False, assuming the year is not a leap year

    # Write your logic here
    if year%4 == 0 and year%100 != 0:       # If the year is divisible by 4 and not divisible by 100, it is a leap year
        leap = True
    if year%100 == 0 and year%400 == 0:     # If the year is divisible by 100 and also divisible by 400, it is also a leap year
        leap = True
    return leap                             # Return the result (True if is a leap year, False otherwise)

    year = int(input())
    print(is_leap(year))
```

**3.1** sWAP cASE

```python
def swap_case(s):
    sModified = ""                  # Initialize an empty string "sModified" to hold the resulting swapped case characters

    for i in s:                     # Iterate over each character in the input string "s"
        if i.isupper():             # If the character is uppercase ...
            sModified += i.lower()    # ... convert it to lowercase and add it to "sModified"

        elif i.islower():           # If the character is lowercase ...
            sModified += i.upper()    # ... convert it to uppercase and add it to "sModified"

        else:                       # If the character is neither uppercase nor lowercase ...
            sModified += i            # ... keep it as is

    return sModified

if __name__ == '__main__':
    s = input()
    result = swap_case(s)
    print(result)
```

```python
def count_substring(string, sub_string):
    cont = 0                                                # Initialize a counter to zero to keep track of the occurrences of "sub_string" in "string"

    for i in range(len(string)-len(sub_string)+1):          # Iterate over the entire length of the "string" - len(sub_string) + 1
        if string[i:i + len(sub_string)] == sub_string:     # Extract a substring of length equal to "sub_string" starting from index "i"
                                                            # if this extracted substring matches "sub_string" ...
            cont += 1                                       # ... increment the counter

    return cont


if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()

    count = count_substring(string, sub_string)
    print(count)
```

**3.10** String Formatting

```python
def print_formatted(number):
    # your code goes here
    width = len(bin(number)) - 2                           # Determine the width required to align all columns based on the binary representation of "number"
                                                           # To get the actual width, subtract 2 to ignore the "0b" prefix

    for i in range(1, number+1):                           # Iterate from 1 to "number" (inclusive because we consider number+1)
        decimal = str(i).rjust(width)                      # Convert the current number "i" to decimal (as a string) and right justify to "width" for alignment
        octal = oct(i)[2:].rjust(width)                    # Convert "i" to octal, remove the "0o" prefix using [2:], and right justify it to "width"
        hexadecimal = hex(i)[2:].upper().rjust(width)      # Convert "i" to hexadecimal, remove the "0x" prefix using [2:], convert to uppercase, and right justify
        binary = bin(i)[2:].rjust(width)                   # Convert "i" to binary, remove the "0b" prefix using [2:], and right justify it to "width"

        print(decimal, octal, hexadecimal, binary)


if __name__ == '__main__':
    n = int(input())
    print_formatted(n)
```

**3.11** Alphabet Rangoli

```python
def print_rangoli(size):
    import string                                                                   # Import the "string" module to get all lowercase English letters
    letters = string.ascii_lowercase                                                # Create a string containing all lowercase letters from "a" to "z"
    rows = []                                                                       # Initialize an empty list to store each row of the rangoli

    for i in range(size):                                                           # Range for the upper half of the rangoli
       left_part = "-".join(letters[size-1:size-i-1:-1] + letters[size-i-1:size])   # Create the left part of the row by selecting and joining letters
                                                                                    # The slice "letters[size-1:size-i-1:-1]" selects letters in reverse from "size-1" to "size-i-1"
                                                                                    # The slice "letters[size-i-1:size]" selects letters from "size-i-1" to "size" in normal order
       rows.append(left_part.center(4*size-3,"-"))                                  # Center the row to fit within the rangoli, with a total width of "4*size - 3"
                                                                                    # Use "-" as the fill character to create symmetrical rows

    complete_rangoli = "\n".join(rows+rows[-2::-1])                                 # Combine the top half and bottom half to form the complete rangoli
                                                                                    # The bottom half is a reverse of the rows excluding the middle row

    print(complete_rangoli)


if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```

**3.13** The Minion Game

```python
def minion_game(string):
    # your code goes here
    vowels = "AEIOU"                          # Define the set of vowels
    stuart_score = 0                          # Initialize Stuart's score
    kevin_score = 0                           # Initialize Kevin's score
    l = len(string)                           # Length of the input string

    for i in range(l):                        # Iterate through each character of the string
        if string[i] in vowels:               # If the current character is a vowel, add points to Kevin's score
            kevin_score += l-i                # Kevin scores all substrings starting from index "i" to the end of the string

        else:                                 # Otherwise, if the current character is a consonant, add points to Stuart's score
            stuart_score += l-i               # Stuart scores all substrings starting from index "i" to the end of the string

    if kevin_score > stuart_score:
        print(f"Kevin {kevin_score}")

    elif stuart_score > kevin_score:
        print(f"Stuart {stuart_score}")

    elif kevin_score == stuart_score:
        print("Draw")


if __name__ == '__main__':
    s = input()
    minion_game(s)
```

```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
T = int(input())                                          # Read an integer input "T" representing the number of test cases

for _ in range(T):                                        # Iterate over each test case
    n = int(input())                                      # Read an integer input "n" representing the number of side lengths of cubes
    side_lengths = list(map(int, input().split()))        # Read the side lengths of the cubes and convert them into a list "side_lengths"
    left = 0                                              # Initialize pointers for left ...
    right = n - 1                                         # ... and right ends of the list
    currentMax = float("inf")                             # Set the current maximum side length that can be chosen (start with infinity, as we will try to maintain decreasing order)

    while left <= right:                                  # Iterate until the left pointer surpasses the right pointer
        if side_lengths[left] >= side_lengths[right]:     # Check if the cube at the left is larger or equal to the cube at the right ...
            chosen = side_lengths[left]                   # Choose the left cube ...
            left += 1                                     # ... and move the left pointer to the right

        else:                                             # Else ...
            chosen = side_lengths[right]                  # Choose the right cube ...
            right -= 1                                    # ... and move the right pointer to the left

        if chosen > currentMax:                           # Check if the chosen cube exceeds the current maximum allowed side length ...
            print("No")                                   # Print "No" ...
            break                                         # ... and exit the loop for this test case

        currentMax = chosen                               # Update the current maximum allowed side length to be the chosen cube

    else:                                                 # If the while loop completes without breaking ...
        print("Yes")                                      # ... print "Yes"
```

**6.2** Time Delta

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the time_delta function below.
import time

def time_delta(t1, t2):
    format_str = "%a %d %b %Y %H:%M:%S"                               # Define the format of the input date and time strings (without the timezone part)

    time1_str, tz1 = t1[:-6], t1[-5:]                                 # Split the input strings into time components and timezone components
                                                                      # "t1[:-6]" excludes the last 6 characters (timezone), "t1[-5:]" gets the last 5 characters
    time2_str, tz2 = t2[:-6], t2[-5:]                                 # The same as before

    time1_tuple = time.strptime(time1_str, format_str)                # Parse the time components to get structured time objects (time tuples)
    time2_tuple = time.strptime(time2_str, format_str)

    time1_epoch = time.mktime(time1_tuple)                            # Convert time tuples to seconds since the epoch (January 1, 1970)
    time2_epoch = time.mktime(time2_tuple)

    tz1_seconds = int(tz1[:3]) * 3600 + int(tz1[0] + tz1[3:]) * 60    # Calculate the timezone offset in seconds for both times
                                                                      # "tz1[:3]" gets the hour part of the timezone, "tz1[3:]" gets the minute part
                                                                      # Convert the timezone to seconds by multiplying hours by 3600 and adding the minutes multiplied by 60
    tz2_seconds = int(tz2[:3]) * 3600 + int(tz2[0] + tz2[3:]) * 60    # The same as before

    time1_corrected = time1_epoch - tz1_seconds                       # Correct the epoch times by subtracting the timezone offsets
    time2_corrected = time2_epoch - tz2_seconds

    return str(int(abs(time1_corrected - time2_corrected)))           # Calculate the absolute difference between the two corrected times

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    t = int(input())

    for t_itr in range(t):
        t1 = input()

        t2 = input()

        delta = time_delta(t1, t2)

        fptr.write(delta + '\n')

    fptr.close()
```

```python
import xml.etree.ElementTree as etree

maxdepth = 0                                          # Define a global variable "maxdepth" to keep track of the maximum depth of the XML tree

def depth(elem, level=1):
    global maxdepth                                   # Declare the "maxdepth" as global to modify it inside the function
    if level + 1> maxdepth:                           # If the current level is greater than maxdepth ...
        maxdepth = level + 1                          # ... update the maximum depth

    for child in elem:                                # Recursively call the depth function on each child element
        depth(child, level + 1)


if __name__ == '__main__':
    n = int(input())
    xml = ""
    for i in range(n):
        xml =  xml + input() + "\n"
    tree = etree.ElementTree(etree.fromstring(xml))
    depth(tree.getroot(), -1)
    print(maxdepth)
```

```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'insertionSort2' function below.
#
# The function accepts following parameters:
#  1. INTEGER n
#  2. INTEGER_ARRAY arr
#

def insertionSort2(n, arr):
    # Write your code here
    for currentValueIndex in range(1, n):                                       # Iterate from the second element to the end of the array
        currentValue = arr[currentValueIndex]                                   # Store the current value that needs to be inserted in the sorted portion

        for currentPosition in range(currentValueIndex -1, -2, -1):             # Iterate backwards in the sorted part to find the correct position for the current value
            if currentPosition >= 0 and arr[currentPosition] > currentValue:    # If the current position is valid and the element is greater than the value being sorted
                arr[currentPosition + 1] = arr[currentPosition]                 # Shift the current element to the right

            else:                                                               # Else ...
                arr[currentPosition + 1] = currentValue                         # ... insert the current value at the correct position
                break

        print(" ".join(map(str, arr)))                                          # Print the array after each iteration of insertion to visualize the sorting process

if __name__ == '__main__':
    n = int(input().strip())

    arr = list(map(int, input().rstrip().split()))

    insertionSort2(n, arr)
```
